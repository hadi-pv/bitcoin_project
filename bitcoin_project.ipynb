```python
import pandas as pd
```

```python
df_train = pd.read_csv("train.csv")
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Purpose_Of_Loan', OneHotEncoder(), [5])], remainder = 'passthrough')
```

```python
X=df_train.drop(['Interest_Rate','Loan_ID'],axis='columns')
X_train=X.to_numpy()
X_train2 = ct.fit_transform(X_train)
X_train2 = pd.DataFrame(X_train2[:,1:])
y_train2=df_train.Interest_Rate
```

```python
model.fit(X_train2,y_train2)
```

```python
df_test=pd.read_csv("test.csv")
X_test=df_train.drop(['Interest_Rate','Loan_ID'],axis='columns').to_numpy()
X_test2 = ct.fit_transform(X_test)
X_test2 = pd.DataFrame(X_test2[:,1:])
X_test2
```

```python
y_predicted=pd.DataFrame(model.predict(X_test2),columns=['Interest_Rate'])
y_predicted
```

```python
final=pd.concat([df_test,y_predicted],axis="columns")
final[['Loan_ID','Interest_Rate']].to_csv("Submission2.csv")
```
